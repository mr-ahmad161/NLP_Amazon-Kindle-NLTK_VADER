# Text Preprocessing

```python
from nltk.corpus import stopwords
import string
stop = set(stopwords.words('english'))
punctuation = list(string.punctuation)
stop.update(punctuation)
```

```python
def clean_text(text):  # function for data cleaning 
    final_text = []
    for i in text.split():
        if i.strip().lower() not in stop:
            word = i.strip().lower()
            final_text.append(word)
    return " ".join(final_text) 
```

```python
def review_sentiment(rating): 
    a="neg"
    b="pos"
    # Replacing rating of 1,2,3 with 0(not good) and 4,5 with 1(good) 
    if(rating == 1 or rating == 2 ):
        return a
    else:
        return b
```

```python
df.overall = df.overall.apply(review_sentiment)
df.head()
```

```python
df['reviewText'] = df['reviewText'].apply(clean_text)
```

```python
sid.polarity_scores(df.loc[0]['reviewText'])
```

```python
df['scores'] = df['reviewText'].apply(lambda reviewText: sid.polarity_scores(reviewText))
df.head()
```

```python
df['compound'] = df['scores'].apply(lambda score_dict: score_dict['compound'])
df.head(10)
```

```python
df['comp_score'] = df['compound'].apply(lambda c: 'pos' if c >= 0.50 else 'neg')
df.head(50)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(classification_report(df['overall'],df['comp_score']))
```

```python
print(confusion_matrix(df['overall'],df['comp_score']))
```

```python
accuracy_score(df['overall'],df['comp_score'])
```
